**Data Preprocessing**

```python
#removing punctuations
import string
def remove_punctuation(text):
  punct_removed="".join([i for i in text if i not in string.punctuation])
  return punct_removed
```

```python
#tokenizing
import nltk
nltk.download("punkt")
def tokenization(text):
  token=nltk.word_tokenize(text)
  return token
```

```python
#removing stopwords
nltk.download('stopwords')
stopword = nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
  output = [i for i in text if i not in stopword]
  return output
```

```python
#lemmatizing
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemm = WordNetLemmatizer()
def lemmatize(text):
  lemm_text = [lemm.lemmatize(word) for word in text]
  return lemm_text
```

```python
# defining a function to carryout all pre-processing techinique on each rows in the "text" column on the data
def preprocess(data_col):
  corpus = []
  for text in data_col:
    new_text = remove_punctuation(text)
    new_text = new_text.lower()
    new_text = tokenization(new_text)
    new_text = remove_stopwords(new_text)
    new_text = lemmatize(new_text)
    corpus.append(' '.join(str(x) for x in new_text))
  return corpus
```

```python
corpus=preprocess(data["text"])
corpus
```

**Model Fitting**

```python
#applying TF-IDF vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer()
train_data = tv.fit_transform(corpus)

X = train_data
y = data['target']
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X,y)
y_pred=lr.predict(X)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y,y_pred)
cm

# Plot confusion matrix
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
acc=accuracy_score(y,y_pred)
acc
```

```python
#defining a function to check the given news is fake or true
def find_news(input):
  input = tv.transform(preprocess(input))
  pred = lr.predict(input)
  if pred ==0:
    print('Fake news')
  else:
    print("True news")
```
